### Посчитаем для каждой книги количество обзоров и среднюю оценку

```python
query_rating = ''' 

SELECT 
    b.title AS book_name,
    b.book_id AS id,
    rew_new.new_reviews AS reviews,
    r_new.avg_rating AS avg_rating
FROM books AS b LEFT JOIN 
                (SELECT
                    book_id,
                    AVG(rating) AS avg_rating
                FROM
                    ratings 
                GROUP BY 
                    book_id
                ) AS r_new ON r_new.book_id = b.book_id
        LEFT JOIN
                (SELECT 
                    book_id,
                    COUNT(review_id) AS new_reviews
                FROM
                    reviews
                GROUP BY 
                    book_id
                ) AS rew_new ON rew_new.book_id = b.book_id
                
; ''' 

request_rating = pd.io.sql.read_sql(query_rating, con = engine)
request_rating.head()
```

```python
request_rating['id'].count()
```

* Составили таблицу с указанием количества обзоров и средней оценкой для каждой книги

### Определим издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
query_publisher = ''' 
SELECT
    S.publisher_id,
    S.publisher,
    S.books_count
FROM
    (SELECT 
        b.publisher_id AS publisher_id,
        COUNT(b.book_id) AS books_count,
        p.publisher AS publisher 
    FROM 
        books AS b
        LEFT JOIN publishers AS p ON b.publisher_id=p.publisher_id

    WHERE 
        b.num_pages > 50
    GROUP BY
        b.publisher_id,
        p.publisher
    ORDER BY 
        books_count DESC) AS S
LIMIT 1
; ''' 

request_publisher = pd.io.sql.read_sql(query_publisher, con = engine)
request_publisher
```

* Наибольшее чисо книг (42 шт.) толще 50 страниц выпустило издательство Penguin Books

### Определим автора с самой высокой средней оценкой книг 

Учитываются только книги с 50 и более оценками

```python
query_author = ''' 
SELECT 
    author_rating.author,
    AVG(author_rating.avg_rating) AS avg_rating
FROM
        (SELECT 
            a.author AS author, 
            AVG(r.rating) AS avg_rating
        FROM books AS b
            LEFT JOIN authors AS a ON b.author_id = a.author_id
            LEFT JOIN ratings AS r ON b.book_id = r.book_id
        GROUP BY 
             b.book_id,a.author_id
        HAVING
            COUNT(r.rating_id) >=50 
        ) AS author_rating
GROUP BY    
    author_rating.author
ORDER BY
    avg_rating DESC
LIMIT 1
    
; ''' 

author = pd.io.sql.read_sql(query_author, con = engine)
author
```

* Автор с самой высокой средней оценкой книг - J.K. Rowling с иллюстрациями Mary GrandPré
* Средний рейтинг 4,28

### Посчитаем среднее количество обзоров от пользователей, которые поставили больше 50 оценок

```python
query_reviews = ''' 

SELECT
    AVG(r.count_reviews)

FROM
    (SELECT username,
            COUNT(review_id) AS count_reviews
    FROM reviews
    WHERE username IN 
        
        (SELECT username
         FROM ratings
         GROUP BY username
            HAVING COUNT(rating_id) > 50)
    GROUP BY username
    ) AS r
; '''

reviews = pd.io.sql.read_sql(query_reviews, con = engine)
reviews
```

* Среднее количество обзоров от пользователей, которые поставили более 50 оценок: 24,3
